```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV
```

## Linear regression

$\mathcal X=\mathbb R^d$, $\mathcal Y=\mathbb R$,
$$ \|w\|_1:=\sum_{i=1}^d |w_i|,\quad \|w\|_2:=\sqrt{\sum_{i=1}^d |w_i|^2},$$
$$\mathcal H=\{x\mapsto h(x)=\langle w,x\rangle+b\},$$
+ Ordinary linear regression
$$ L_S(h)=\frac{1}{m}\sum_{i=1}^m(\langle w,x_i\rangle+b-y_i)^2\to\min_{w,b}$$
+ Ridge regression
$$ L_S(h)+\lambda\|w\|_2^2\to\min_{w,b}$$
+ Lasso
$$ L_S(h)+\lambda\|w\|_1\to\min_{w,b}$$
+ Elastic net
$$ L_S(h)+\lambda_1\|w\|_2^2+\lambda_2\|w\|_1\to\min_{w,b}$$

Assume that samples correspond to the values of a 3rd order polynomial perturbed by a Gaussian noise:
$$ y_i=x_i^3-5 x_i^2+3 x_i+1+\xi_i,\quad \xi_i\sim N(0,1),\quad x_i\sim U(-1,5)$$
$i\in\{1,\dots,m\}$, $m=30$.

**1. Plot the graph of the unperturbed function and the scatter plot of $(x_i,y_i)$.**

```python
def polynom(x, eps):
    return x * x * x - 5 * x * x + 3 * x + 1 + eps
```

```python
m = 35
```

```python
np.random.seed(23)
eps = np.random.normal(0, 1, size=m)
x = np.random.uniform(-1, 5, size=(m,))
x_tmp = np.linspace(-1, 5)
```

```python
plt.scatter(x, polynom(x, eps))
plt.plot(x_tmp, polynom(x_tmp, 0), color='r')
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

### Ordinary linear regression
Consider an extended set of features: $\mathbf{x}=(x,x^2,\dots,x^d)$ (`PolynomialFeatures`) and the correspondent linear regression problem:
$$ L_S(h)=\frac{1}{m}\sum_{i=1}^m(\langle\mathbf{w},\mathbf{x}_i\rangle+b-y_i)^2\to\min_{\mathbf{w},b}$$
The hypothesis class
$$\mathcal H_d=\{\mathbf{x}\mapsto h(\mathbf{x})=\langle\mathbf{w},\mathbf{x}\rangle+b\},$$
depends on the hyperparameter $d$. Consider the training errors
$$ \psi(d)=\inf_{\mathbf{w},b} L_S(h)$$
and average cross-validation errors 
$$ \overline\psi(d)=\frac{1}{k}\sum_{i=1}^k\psi_i(d),\quad \psi_i(d)=\inf_{\mathbf{w},b} L_{S\backslash S_i}(h)$$
for different values of $d$. Here $(S_1,\dots,S_d)$ is a partition of $S$, considered in the cross-validation. 

**2. Find $d$ with the best average cross-validation score.** 

```python
np.random.seed(23)
m = 35
x = np.random.uniform(-1, 5, size=(m,))
x_train = x.reshape(-1, 1)
eps = np.random.normal(0, 1, size=m)
y_train = polynom(x, eps)
```

```python
poly_degree = []
train_errors = []
cv_errors = []
for i in range(0, 15):
    poly_degree.append(i)
    poly_lr = make_pipeline(PolynomialFeatures(degree=i), LinearRegression())
    poly_lr.fit(x_train, y_train)
    train_errors.append(mean_squared_error(y_train, poly_lr.predict(x_train)))

for i in range(0, 15):
    poly_lr = make_pipeline(PolynomialFeatures(degree=i, include_bias=True), LinearRegression())
    scores = cross_val_score(poly_lr, x_train, y_train, scoring='neg_mean_squared_error', cv=5)
    cv_errors.append(np.mean(scores) * -1)
```

```python
plt.scatter(poly_degree, train_errors, color='b', label='Train error')
plt.scatter(poly_degree, cv_errors, color='r', label='Avg CV error')
plt.xlabel("polynomial degree")
plt.ylabel("errors")
plt.legend()
plt.show()
```

**3. (a) Compute the training error of this predictor for the RMSE (root mean square error) evaluation metrics. Compare with the standard deviation of the data. (b) By the means of scatter plots compare the coefficients of the best predictor and the original coefficients.** 

```python
poly = PolynomialFeatures(degree=3)
lr = LinearRegression()

x_train_poly = poly.fit_transform(x_train)
lr.fit(x_train_poly, y_train)
```

```python
rmse_error = np.sqrt(mean_squared_error(y_train, lr.predict(x_train_poly)))

print('Best degree: 3')
print(f'RMSE for the regression predictor with the best d: {rmse_error}')
print(f'y standard deviation: {np.std(y_train)}')
```

```python
plt.scatter([0, 1, 2, 3], lr.coef_, color='b', alpha=0.5, label='Regression coef')
plt.scatter([0, 1, 2, 3], [1, 3, -5, 1], color='orange', alpha=0.5, label='True coef')
plt.legend()
plt.show()
```

**4. Show the graphs of the predictors for d=3 and d=10 along with the unperturbed original function.**

```python
poly_3 = PolynomialFeatures(degree=3)
lr_3 = LinearRegression()

x_train_poly_3 = poly_3.fit_transform(x_train)
lr_3.fit(x_train_poly_3, y_train)
```

```python
poly_bad = PolynomialFeatures(degree=14)
lr_bad = LinearRegression()

x_train_poly_bad = poly_bad.fit_transform(x_train)
lr_bad.fit(x_train_poly_bad, y_train)
```

```python
x_linspace = np.linspace(-1, 5, 100).reshape(-1, 1)

plt.plot(x_linspace, polynom(x_linspace, 0), color='r', label='f')
plt.plot(x_linspace, lr_3.predict(poly_3.fit_transform(x_linspace)), color='b', label='d = 3')
plt.plot(x_linspace, lr_bad.predict(poly_bad.fit_transform(x_linspace)), color='orange', label='d = 14')
plt.legend()
```

### Ridge regression
Objective function in scikit-learn:
$$ m\cdot L_S(h)+\alpha\|w\|_2^2$$
**5. (a) Take a class polynomials, having a fixed large degree ($d=10$) and select the best regularization parameter $\alpha$. Compute the training error for the RMSE evaluation metrics for the best 
$\alpha$.**  

Consider log-scale for $\alpha$:
`alphas=np.logspace(a_1,a_2,N=50,endpoint=True,b=10.0)`: 
$$ z_i=a_1+\frac{a_2-a_1}{N-1}i,\quad i=0,\dots,N-1,$$
$$ \alpha_i=b^{z_i},\quad i=0,\dots,N-1.$$
Note that the difference
$$ \log_b \alpha_{i+1}-\log_b(\alpha_i)=z_{i+1}-z_i=\frac{a_2-a_1}{N-1}$$
is constant.

**5. (b) Show the dependence of the training and cross-validation errors on the regularization parameter**

```python
poly = PolynomialFeatures(degree=14)
x_train_poly = poly.fit_transform(x_train)
```

```python
alphas = np.logspace(0, 4, num=100, endpoint=True, base=10.0)
```

```python
train_errors = []
cv_errors = []

for alpha in alphas: 
    rr = Ridge(alpha=alpha)
    rr.fit(x_train_poly, y_train)
    train_errors.append(mean_squared_error(y_train, rr.predict(x_train_poly)))
    
for alpha in alphas: 
    rr = Ridge(alpha=alpha)
    scores = cross_val_score(rr, x_train_poly, y_train, scoring='neg_mean_squared_error', cv=5)
    cv_errors.append(np.mean(scores) * -1)
```

```python
plt.scatter(alphas, train_errors, color='b', label='Train error')
plt.scatter(alphas, cv_errors, color='orange', label='Avg CV error')
plt.xlabel("Regularization parameter alpha")
plt.ylabel("errors")
plt.legend()
plt.show()
```

```python
best_alpha = alphas[np.argmin(cv_errors)]
```

```python
rr = Ridge(alpha=best_alpha)
rr.fit(x_train_poly, y_train)
rmse_error = np.sqrt(mean_squared_error(y_train, rr.predict(x_train_poly)))

print(f'Best alpha: {best_alpha}')
print(f'RMSE of Ridge regression with best alpha: {rmse_error}')
```

**6. Show the graphs of the best Ridge predictor along with the graph of the unperturbed original function.**

```python
x_linspace = np.linspace(-1, 5, 100).reshape(-1, 1)

plt.plot(x_linspace, polynom(x_linspace, 0), color='r', label='f')
plt.plot(x_linspace, rr.predict(poly.fit_transform(x_linspace)), color='b', label=f'alpha = {best_alpha}')
plt.legend()
```

### Lasso regression
Objective function in scikit-learn:
$$ \frac{1}{2}L_S(h)+\alpha\|w\|_1.$$
**7. The problem is the same as 5, 6 for Ridge regression.**

```python
poly = PolynomialFeatures(degree=14)
x_train_poly = poly.fit_transform(x_train)
```

```python
alphas = np.logspace(-10, 1, num=100, endpoint=True, base=10.0)
```

```python
train_errors = []
cv_errors = []

for alpha in alphas: 
    ls = Lasso(alpha=alpha)
    ls.fit(x_train_poly, y_train)
    train_errors.append(mean_squared_error(y_train, ls.predict(x_train_poly)))
    
for alpha in alphas: 
    ls = Lasso(alpha=alpha)
    scores = cross_val_score(ls, x_train_poly, y_train, scoring='neg_mean_squared_error', cv=5)
    cv_errors.append(np.mean(scores) * -1)
```

```python
plt.scatter(alphas, train_errors, color='b', label='Train error')
plt.scatter(alphas, cv_errors, color='orange', label='Avg CV error')
plt.xlabel("Regularization parameter alpha")
plt.ylabel("errors")
plt.legend()
plt.show()
```

```python
best_alpha = alphas[np.argmin(cv_errors)]
```

```python
ls = Lasso(alpha=best_alpha)
ls.fit(x_train_poly, y_train)
rmse_error = np.sqrt(mean_squared_error(y_train, ls.predict(x_train_poly)))

print(f'Best alpha: {best_alpha}')
print(f'RMSE of Lasso regression with best alpha: {rmse_error}')
```

```python
x_linspace = np.linspace(-1, 5, 100).reshape(-1, 1)

plt.plot(x_linspace, polynom(x_linspace, 0), color='r', label='f')
plt.plot(x_linspace, ls.predict(poly.fit_transform(x_linspace)), color='b', label=f'alpha = {best_alpha}')
plt.legend()
```

## Elastic net
Objective function in scikit-learn: 
$$ \frac{1}{2}L_S(h)+\alpha r \|\mathbf w\|_1+\frac{1}{2}\alpha(1-r)\|\mathbf{w}\|_2^2.$$
**8. Fix some large $d$ and use `GridSearchCV` to find the best combination of $\alpha$ and $r$ (`l1_ratio`).**

```python
poly = PolynomialFeatures(degree=14)
x_train_poly = poly.fit_transform(x_train)
```

```python
elastic_net = ElasticNet()
```

```python
param_grid = {'alpha': np.logspace(-10, 2, 50),
              'l1_ratio': np.linspace(0.1, 0.9, 50)}

grid_search = GridSearchCV(estimator=elastic_net, param_grid=param_grid, scoring='neg_mean_squared_error', cv=5)

grid_search.fit(x_train_poly, y_train)

print("Best parameters: ", grid_search.best_params_)
print("Best estimator: ", grid_search.best_estimator_)
print("RMSE for the best papameters: ", np.sqrt(grid_search.best_score_ * -1))
```

**9. Show the graphs of the best predictor along with the graph of the unperturbed original function.**

```python
elastic_net = ElasticNet(alpha=grid_search.best_params_['alpha'], l1_ratio=grid_search.best_params_['l1_ratio'])
elastic_net.fit(x_train_poly, y_train)
```

```python
x_linspace = np.linspace(-1, 5, 100).reshape(-1, 1)

plt.plot(x_linspace, polynom(x_linspace, 0), color='r', label='f')
plt.plot(x_linspace, elastic_net.predict(poly.fit_transform(x_linspace)), color='b', label=f'Elastic net')
plt.legend()
```
