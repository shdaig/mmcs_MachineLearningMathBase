We are interested in algorithms that can learn from data.

* Examples (dataset): $S=(x_i,y_i)_{i=1}^n$:
* $x_i\in\mathcal X=\mathbb R^d$: instances, features, inputs,
* $y_i\in\mathcal Y$: labels, responses.


Examples
* $S$ is a set of houses; features: the number of rooms, distance to the center, crime rate, ...; labels: house prices;
* $S$ is a set of e-mails; features: words in the email; labels: spam/non-spam;
* $S$ is a set of images of cats and dogs; features: $k\times m$ matrices of integer pixel intensities from 0 (white) to 256 (black); labels: cat/dog.
* recommender systems, automatic stock trading, automatic translation, cancer (or Covid-19) detection, data driven discoveries in science

**Supervised learing**

Given i.i.d. examples:

$$(x_i,y_i)\sim\mathcal P,\quad (x_i,y_i)\in\mathbb R^d\times\mathbb R$$

we want to find $\hat f: \hat f(x_i)\approx y_i$ 

with good generalization properties: $\hat f(x)\approx y$ for unseen examples $(x,y)$.

Two types of the supervised learning:
* Regression: $\mathcal Y=\mathbb R$; $y$ is quantitative (continuous)
* Classification: $\mathcal Y$ is finite; $y$ is qualitative (discrete)

There is a natural metrics on $\mathbb R$, but there is no natural distance between classes.

**Nearest neighbors**
$$\mathcal N_k(x):=\textrm{ the set of $k$ points $x_i$ closest to $x$}. $$
* k-NN for regression problem:
$$\hat f(x)=\frac{1}{k}\sum_{i:x_i\in\mathcal N_k} y_i:\quad \textrm{averaging}.$$
* k-NN for classification problem:
$$ \hat f(x)=\arg\max_{y\in\mathcal Y}\frac{1}{k}\sum_{i:x_i\in\mathcal N_k(x)} I_{\{y_i=y\}}\quad \textrm{majority vote}.$$

**1  neighbor**

```python
from sklearn.metrics import euclidean_distances
from sklearn.neighbors import KNeighborsClassifier

def plot_n_neighbors(X, y, n_neighbors=1):
    X_test = np.array([[0.5, 3], [4, 2], [-.5, .5]])
    # `euclidean_distances`: considering the rows of X and Y as vectors, computes the distance 
    #  matrix between each pair of vectors
    # Returns: distances ndarray of shape (n_samples_X, n_samples_Y)
    dist = euclidean_distances(X, X_test)
    # for each test point (column) sorts the train points in distance ascending order
    closest = np.argsort(dist, axis=0)
    plt.figure(figsize=(9,7))
  # plt.gca().set_aspect("equal")
    for x, neighbors in zip(X_test, closest.T):
        for neighbor in neighbors[:n_neighbors]:
            plt.arrow(x[0], x[1], X[neighbor, 0] - x[0],
                    X[neighbor, 1] - x[1], head_width=0, fc='k', ec='k')
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)
    test_points = plt.scatter(X_test[:, 0], X_test[:, 1], c=clf.predict(X_test), marker="*")
    training_points = plt.scatter(X[:, 0], X[:, 1], c=y)    
plot_n_neighbors(X, y)
plt.savefig("images/knn_boundary_k1.png", bbox_inches='tight')
```

**Decision boundary** for different number of neighbors.

```python
# compute and show the predictions on a large grid
def plot_2d_classification(classifier, X, fill=False, ax=None, eps=None, alpha=1):       
    # multiclass                                                                                                                 
    if eps is None:                                                                                                              
      eps = X.std() / 2.                                                                                                       

    if ax is None:                                                                                                               
      ax = plt.gca()                                                                                                           

    x_min, x_max = X[:, 0].min() - eps, X[:, 0].max() + eps                                                                      
    y_min, y_max = X[:, 1].min() - eps, X[:, 1].max() + eps                                                                      
    xx = np.linspace(x_min, x_max, 1000)                                                                                         
    yy = np.linspace(y_min, y_max, 1000)
    
    # X1[i,:]=xx, X2[:,i]=yy
    # arrays of projections of two-dimensional points to the axes
    X1, X2 = np.meshgrid(xx, yy)
    # ravel: flattened array; 
    # X_grid: Cartesian product of xx, yy
    X_grid = np.c_[X1.ravel(), X2.ravel()]                                                                                       
    decision_values = classifier.predict(X_grid)                                                                                 
    # Display data as an image
    ax.imshow(decision_values.reshape(X1.shape), extent=(x_min, x_max,                                                           
                                                       y_min, y_max),                                                          
            aspect='auto', origin='lower', alpha=alpha)                                                               
    ax.set_xlim(x_min, x_max)                                                                                                    
    ax.set_ylim(y_min, y_max)                                                                                                    
    ax.set_xticks(())                                                                                                            
    ax.set_yticks(())    
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 9))
for ax, n_neighbors in zip(axes.ravel(), [1, 5, 10, 30]):
    ax.set_title(f"n_neighbors={n_neighbors}")
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)
    ax.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k')
    plot_2d_classification(clf, X, ax=ax, alpha=.5)
   # ax.set_aspect("equal")
plt.savefig("images/knn_boundary_varying_k.png", bbox_inches='tight')
```

The larger the number of neighbors, the simpler the model and the decision boundary. The reason: more averaging.

**Test scores and model selection**

Split the data set $S$ into training and validation (testing) parts: $S=T\cup V$, $T\cap V=\emptyset$. 

Fit the model on $T$ and get $\hat f$. 

Compute the accuracy score on the validation part:

$$ {\tt test}\underline{ }{\tt score}=\frac{1}{|V|}\sum_{(x_i,y_i)\in V} I_{\{\hat f(x_i)= y_i\}}.$$

```python
neighbors = range(1, 30, 2)

training_scores = []
test_scores = []
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=13)

for n_neighbors in neighbors:
    knn = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X_train, y_train)
    training_scores.append(knn.score(X_train, y_train))
    test_scores.append(knn.score(X_test, y_test))
```

```python
plt.figure(figsize=(9, 7))
plt.plot(neighbors, training_scores, label="training scores")
plt.plot(neighbors, test_scores, label="test scores")
plt.ylabel("accuracy")
plt.xlabel("n_neighbors")
plt.legend()
plt.savefig("images/knn_model_complexity.png", bbox_inches='tight')
```

Model selection: take the model with the best test score, which is the estimate of the true score on the fresh data.

**Cross validation**

Cross validation is a method for the estimation of the true score. 

Split $S$ into $K$ subsets $S_1,\dots,S_K$ at random.

Denote by $\hat f^{(-k)}$ the  model $\hat f$ fitted on $S\backslash S_k$. 

The cross-validation estimate of the true accuracy score of $\hat f$:
$$ \textrm{CV}(\hat f)=\frac{1}{K}\sum_{k=1}^K\frac{1}{|S_k|}\sum_{i\in S_k} I_{\{\hat f^{(-k)}(x_i)= y_i\}}.$$

Compared to the validation method considered above, 
the cross-validation 
* uses data more effectively
* is more stable

**Cross-validation + test set**

```python
Image(filename = "Cross_validation_and_test_set.png", width = 800, height = 400)
```

```python
from sklearn.model_selection import cross_val_score

X_train, X_test, y_train, y_test = train_test_split(X, y)

cross_val_scores = []

for i in neighbors:
    knn = KNeighborsClassifier(n_neighbors=i)
    scores = cross_val_score(knn, X_train, y_train, cv=10)
    cross_val_scores.append(np.mean(scores))
    
print(f"best cross-validation score: {np.max(cross_val_scores):.3}")
best_n_neighbors = neighbors[np.argmax(cross_val_scores)]
print(f"best n_neighbors: {best_n_neighbors}")

knn = KNeighborsClassifier(n_neighbors=best_n_neighbors)
knn.fit(X_train, y_train)
print(f"test-set score: {knn.score(X_test, y_test):.3f}")
```

 **Leave-One-Out Cross-Validation (LOOCV)**

Each fold is a single sample: $K=n$,
$$ \textrm{CV}(\hat f)=\frac{1}{n}\sum_{k=1}^n I_{\{\hat f^{(-k)}(x_k)= y_k\}}.$$

```python
from sklearn.model_selection import LeaveOneOut
loo = LeaveOneOut()
scores = cross_val_score(knn, iris.data, iris.target, cv=loo)
print("Number of folds: ", len(scores))
print("Mean accuracy: {:.2f}".format(scores.mean()))
```

**Nearest neighbors for regression**

```python
X=np.linspace(0,10,100)
y = np.sin(X)+np.random.normal(0,1,np.shape(X)[0])
plt.plot(y)
```

```python
from sklearn.neighbors import KNeighborsRegressor
X_train, X_test, y_train, y_test = train_test_split(X.reshape(-1,1), y, random_state=0)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
# create 1,000 data points, evenly spaced between -3 and 3
line = np.linspace(0, 10, 100).reshape(-1, 1)
for n_neighbors, ax in zip([1, 3, 9], axes):
    # make predictions using 1, 3, or 9 neighbors
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^')
    ax.plot(X_test, y_test, 'v')

    ax.set_title(
        "{} neighbor(s)\n train score: {:.2f} test score: {:.2f}".format(
            n_neighbors, reg.score(X_train, y_train),
            reg.score(X_test, y_test)))
    ax.set_xlabel("Feature")
    ax.set_ylabel("Target")
axes[0].legend(["Model predictions", "Training data/target",
                "Test data/target"], loc="best")
```

Using only a single neighbor, each point in the training set has an obvious influence on the predictions, and the predicted values go through all of the data points. This leads to a very unsteady prediction. Considering more neighbors leads to smoother predictions, but these do not fit the training data as well.

The default score: $R^2$ coefficient of determination:
$$ R^2=1-\frac{\sum_{i=1}^n(y_i-\hat y_i)^2}{\sum_{i=1}^n(y_i-\overline y)^2}$$

```python
param_grid = {'n_neighbors':  np.arange(1, 30, 2)}
grid = GridSearchCV(KNeighborsRegressor(), param_grid=param_grid, cv=10,
                   return_train_score=True)
grid.fit(X_train, y_train)
print(f"best parameters: {grid.best_params_}")

print(f"test-set score: {grid.score(X_test, y_test):.3f}")
```

**The wrong and right way to do cross-validation**

Idea: Hastie et al., 2009; 7.10.2

```python
import numpy as np
from scipy import stats
```

```python
n=50
d=5000
```

```python
np.random.seed(17)
X=np.random.normal(size=(n,d))
X.shape
```

```python
y=np.random.binomial(1,0.5,size=n)
y
```

Correlation coefficient of $x,y$:
$$ r=\frac{\sum_{i=1}^n(x_i-\overline x)(y_i-\overline y)}{\sqrt{\sum_{i=1}^n(x_i-\overline x)^2}\sqrt{\sum_{i=1}^n(y_i-\overline y)^2}}$$
Select 100 features, having the largest correlation with labels:

```python
def best_features(X,y,k,index):
    c=np.zeros(d)
    for i in index:
        c[i], _ = stats.pearsonr(X[:,i],y)
    J=np.argsort(np.abs(c))[d-k:d]
    return J   
np.arange(d)
```

```python
k=100
J=best_features(X,y,k,np.arange(d))
X_best=X[:,J]
J
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

Cross-validation: wrong way

```python
knn=KNeighborsClassifier(n_neighbors=1) 
np.mean(cross_val_score(knn,X_best,y,cv=5))
```

```python
logreg = LogisticRegression()
np.mean(cross_val_score(logreg,X_best,y,cv=5))
```

Test (fresh) examples

```python
X_test=np.random.normal(size=(n,k))
y_test=np.random.binomial(1,0.5,size=n)
```

```python
from sklearn.metrics import accuracy_score
```

```python
knn.fit(X_best,y)
accuracy_score(y_test,knn.predict(X_test))
```

```python
logreg.fit(X_best,y)
accuracy_score(y_test,logreg.predict(X_test))
```

The mistake: model is cross-validated over the same data, which were used in its construction.

Cross-validation: right way

```python
from sklearn.model_selection import KFold
```

```python
k=100
kf = KFold(n_splits=5,shuffle=True,random_state=3)
z=[]
knn=KNeighborsClassifier(n_neighbors=1) 
for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    J=best_features(X,y,k,train_index)
    knn.fit(X_train[:,J],y_train)
    z.append(accuracy_score(y_test,knn.predict(X_test[:,J])))
np.mean(z)
```
