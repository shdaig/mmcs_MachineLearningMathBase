#### Numpy 

Пакет `NumPy` (имя модуля `numpy`) предоставляет строительные блоки для вычислений. Он содержит все стандартные математические функции такие как `sin, cos, tan` и др., а также генераторы случайных чисел, средства линейной алгебры и статистики.

```python
import numpy as np 
x = np.cos(1)
data = [1,2,3,4,5]
y = np.mean(data)
z = np.std(data)
print('cos(1) = {0:1.8f} mean = {1} std = {2}'.format(x,y,z))
```

Основным типом данных в `numpy` является `ndarray`. Этот тип данных позволяет выполнять быстрые матричные операции. Он необходим при работе с большими объемами данных. Объекты `ndarray` можно создавать различными способами. Следующий код создает массив нулей $2\times 3\times 2$.

```python
A = np.zeros([2,3,2]) # 2 by 3 by 2 array of zeros
print(A)
print(A.shape) # number of rows and columns
print(type(A)) # A is an ndarray
```

Можно также использовать метод `range` и списки для создания `ndarray` с помощью метода `array`.

```python
a = np.array(range(4)) # equivalent to np.arange(4)
b = np.array([0,1,2,3])
C = np.array([[1,2,3],[3,2,1]])
print(a, '\n', b,'\n' , C)
```

Размерность объекта `ndarray` можно узнать с помощью метода `shape`, который возвращает кортеж. Форма массива может быть изменена с помощью метода `reshape`. При этом текущий объект не изменяется: требуется создать новый экземпляр.

```python
a = np.array(range(9)) #a is an ndarray of shape (9,)
print(a.shape)
A = a.reshape(3,3) #A is an ndarray of shape (3,3)
print(a)
print(A)
```

В качестве одной из размерностей для `reshape` можно указать $−1$. Тогда эта размерность будет определена исходя из остальных.

Метод `T` вычисляет транспонированную матрицу. Транспонирование ``вектора'' вида `(n,)`  не изменяет его. Чтобы получить строку (столбец) нужно применить метод `reshape`.

```python
a = np.arange(3) #1D array (vector) of shape (3,)
print(a)
print(a.shape)
b = a.reshape(-1,1) # 3x1 array (matrix) of shape (3,1)
print(b)
print(b.T)
A = np.arange(9).reshape(3,3)
print(A.T)
```

Методы `hstack` и `vstack` соединяют массивы горизонтально и вертикально соответственно.

```python
A = np.ones((3,3))
B = np.zeros((3,2))
C = np.hstack((A,B))
print(C)
```

`NumPy` позволяет делать арифметические операции над массивами разных размерностей. Пусть имеются два массива размерностей $(m_1,m_2,\dots,m_p)$ и $(n_1, n_2,\dots, n_p)$ соответственно. Эти массивы называются выровненными (aligned), если
* $m_i = n_i$, или
* $\min\{m_i, n_i\} = 1$, или
* оба значения $m_i$, $n_i$, или одно из них пропущены для всех $ i = 1,\dots, p$. 

Например, массивы вида $(1, 2, 3)$ и $(4, 2, 1)$ выровнены, как и (2, , ) и (1, 2, 3). Пример невыровненных массивов: $(2, 2, 2)$ и $(1, 2, 3)$. `NumPy` ``дублирует'' элементы массива вдоль меньшей размерности до совпадения с большей размерностью. Этот процесс называется трансляцией (broadcasting).

```python
import numpy as np
A= np.arange(4).reshape(2,2) # (2,2) array
x1 = np.array([40,500]) # (2,) array
x2 = x1.reshape(2,1) # (2,1) array
print('A:\n',A,'\nx1:\n',x1,'\nx2:\n',x2)
```

```python
print(A + x1) # shapes (2,2) and (2,)
print(A * x2) # shapes (2,2) and (2,1)
```

Функции, такие как `sum, mean` и `std`, также могут применяться к объектам `ndarray`. Аргумент `axis` может указывать, вдоль какой размерности применяется функция. По умолчанию `axis=None`.

```python
a = np.array(range(4)).reshape(2,2)
print(a)
```

```python
print(a.sum(axis=0)) #summing over rows gives column totals
```

#### Matplotlib

Основной графической библиотекой Python для 2D- и 3D-графики является `matplotlib`, а ее подпакет `pyplot` содержит набор функций, которые делают графику в Python похожей на графику в MATLAB.

Приведенный ниже код иллюстрирует различные возможности создания графиков. Можно изменять стиль и цвет линий и маркеров, а также размер шрифта надписей. 


```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0, 10, 0.1)
u = np.arange(0,10)
y = np.sqrt(x)
v = u/3
plt.figure(figsize = [5,3]) # size of plot in inches
plt.plot(x,y, 'g--') # plot green dashed line
plt.plot(u,v,'r.') # plot red dots
plt.xlabel('x')
plt.ylabel('y')
plt.tight_layout()
# plt.savefig('sqrtplot.pdf',format='pdf') # saving as pdf
plt.show() # both plots will now be drawn
```

Библиотека matplotlib также позволяет создавать подграфики. Ниже на одном рисунке показаны диаграмма рассеяния и гистограмма. При создании гистограммы есть несколько необязательных аргументов, влияющих на расположение графика. Количество корзин гистограммы определяется параметром `bins` (по умолчанию 10). Диаграммы рассеяния также принимают ряд параметров, таких как строка `c`, определяющий цвет точек, и `alpha`, влияющий на прозрачность точек. 

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randn(1000)
u = np.random.randn(100)
v = np.random.randn(100)
plt.subplot(121) # first subplot
plt.hist(x,bins=25, facecolor='b')
plt.xlabel('X Variable')
plt.ylabel('Counts')
plt.subplot(122) # second subplot
plt.scatter(u,v,c='b', alpha=0.5)
plt.show()
```

Можно также создавать трехмерные графики.

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
def npdf(x,y):
    return np.exp(-0.5*(pow(x,2)+pow(y,2)))/np.sqrt(2*np.pi)
x, y = np.random.randn(100), np.random.randn(100)
z = npdf(x,y)
xgrid , ygrid = np.linspace(-3,3,100), np.linspace(-3,3,100)
Xarray , Yarray = np.meshgrid(xgrid ,ygrid)
Zarray = npdf(Xarray ,Yarray)
fig = plt.figure(figsize=plt.figaspect(0.4))
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(x,y,z, c='g')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.set_zlabel('$f(x,y)$')
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(Xarray ,Yarray ,Zarray ,cmap='viridis',
edgecolor='none')
ax2.set_xlabel('$x$')
ax2.set_ylabel('$y$')
ax2.set_zlabel('$f(x,y)$')
plt.show()
```

#### Pandas

Данный модуль предоставляет специальные структуры данных и операции для манипулирования числовыми таблицами и временн\'{ы}ми рядами. Название библиотеки происходит от эконометрического термина ``панельные данные''.  Работа `pandas` с данными строится поверх библиотеки `NumPy`, являющейся инструментом более низкого уровня. 

Две основные структуры данных `pandas` --- это `Series` и `DataFrame`. Объект `Series` можно рассматривать как комбинацию словаря и одномерного `ndarray`. Синтаксис для создания объекта Series: 

`series = pd.Series(<data>, index=['index'])`

Здесь `<data>` — некоторая одномерная структура данных, такая как одномерный `ndarray`, список или словарь, а `index` — это список имен той же длины, что и `<data>`. Когда `<data>` является словарем, индекс создается из ключей словаря. Когда `<data>` — это `ndarray`, а `index` опущен, индексом по умолчанию будет `[0, ..., len(data)-1]`. 

```python
import pandas as pd
DICT = {'one':1, 'two':2, 'three':3, 'four':4}
print(pd.Series(DICT))
```

```python
years = ['2000','2001','2002']
cost = [2.34, 2.89, 3.01]
print(pd.Series(cost ,index = years , name = 'MySeries')) #name it
```

Наиболее часто используемой структурой данных в pandas является двумерный `DataFrame`, который можно рассматривать как реализацию электронной таблицы в pandas или как словарь, в котором каждый `ключ` словаря соответствует имени столбца и `значение` — это данные в этом столбце. Для создания `DataFrame` можно использовать метод `pandas DataFrame`, который имеет три основных аргумента: данные, индекс (метки строк) и столбцы (метки столбцов).

`DataFrame(<data>, index=['<row_name>'], columns=['<column_name>'])`

По умолчанию индексом является `[0, ..., len(data)-1]`. Данные могут быть прочитаны из непосредственно из CSV или Excel файла. Если для создания DataFrame используется словарь, то ключи словаря используются в качестве имен столбцов.

```python
DICT = {'numbers':[1,2,3,4], 'squared':[1,4,9,16] }
df = pd.DataFrame(DICT , index = list('abcd'))
print(df)
```

Часто данные, находящиеся в объектах `DataFrame` или `Series`, необходимо извлечь, изменить или объединить. Получение, настройка и удаление столбцов работает так же, как и для словарей. Следующий код иллюстрирует различные операции. 

```python
ages = [6,3,5,6,5,8,0,3]
d={'Gender':['M', 'F']*4, 'Age': ages}
df1 = pd.DataFrame(d)
df1
```

```python
df1.at[0,'Age']= 60 # change an element
df1.at[1,'Gender'] = 'Female' # change another element
df1
```

```python
df2 = df1.drop('Age',1) # drop a column
df3 = df2.copy(); # create a separate copy of df2
df3['Age'] = ages # add the original column
dfcomb = pd.concat([df1,df2,df3],axis=1) # combine the three dfs
print(dfcomb)
```

#### Scikit-learn
`Scikit-learn` - это библиотека машинного обучения и обработки данных с открытым исходным кодом (имя модуля `sklearn`). Рассмотрим некоторые его возможности.

Предположим, что обучающие данные описываются матрицей признаков $X$ и вектором меток $y$. Следующий код случайным образом разбивает данные на обучающий и тестовый наборы. При этом набор для тестирования составляет половину от общего набора. Классам 0 и 1 соответствуют зеленый и синий цвета.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
np.random.seed(1234)
X=np.pi*(2*np.random.random(size=(400,2))-1)
y=(np.cos(X[:,0])*np.sin(X[:,1]) >=0)
X.shape, y.shape
```

```python
X_train , X_test , y_train , y_test = train_test_split(X, y,
test_size=0.5)
fig = plt.figure()
ax = fig.add_subplot (111)
ax.scatter(X_train[y_train==0,0],X_train[y_train==0,1], c='g',
marker='o',alpha=0.5)
ax.scatter(X_train[y_train==1,0],X_train[y_train==1,1], c='b',
marker='o',alpha=0.5)
ax.scatter(X_test[y_test==0,0],X_test[y_test==0,1], c='g',
marker='s',alpha=0.5)
ax.scatter(X_test[y_test==1,0],X_test[y_test==1,1], c='b',
marker='s',alpha=0.5)
#plt.savefig('sklearntraintest.pdf',format='pdf')
plt.show()
```

Часто требуется стандартизация (масштабирование) данных. В `sklearn` это можно сделать с помощью таких методов как `MinMaxScaler` или `StandardScaler`. Масштабирование улучшает качество работы ряда методов машинного обучения и полезно при визуализации данных. Например, мы хотим стандартизировать $X$ так, чтобы каждое значение находилось между 0 и 1. 

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0, 1))
x_scaled = min_max_scaler.fit_transform(X)
# equivalent to:
x_scaled = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))
```

После того, как данные были разделены и при необходимости стандартизированы, на них могут обучаться модели классификации или регрессии. Общая схема обучения и последующего предсказания в `scikit-learn` выглядит следующим образом. 

`from sklearn.someSubpackage import someClassifier
clf = someClassifier() # choose appropriate classifier
clf.fit(X_train , y_train) # fit the data
y_prediction = clf.predict(X_test) # predict`

В приведенной выше схеме могут быть использованы классификаторы связанные с моделями логистической регрессии, $K$-ближайших соседей, методов опорных векторов и др.

Качество прогнозов модели может быть оценено на основе различных показателей. Например, для задачи классификации можно посмотреть на матрицу ошибок. Следующий код делает это для данных, показанных на рисунке, используя метод опорных векторов. 

```python
from sklearn import svm
clf = svm.SVC(kernel = 'rbf')
clf.fit(X_train , y_train)
y_prediction = clf.predict(X_test)
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test , y_prediction))
```

#### Iris dataset

Рассмотрим несколько более подробно один небольшой классический набор данных. Он содержит результаты измерений цветков ириса, относящихся к одному из трех сортов: setosa, versicolor, или virginica. Данный набор включен в модуль `dataaset` библиотеки `scikit-learn`. Его можно загрузить с помощью функции `load\_iris`.

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

Объект `iris_dataset`, который возвращает `load_iris` является объектом `Bunch`, который похож на словарь. Он содержит ключи и значения.

```python
print("Keys of iris_dataset: \n{}".format(iris_dataset.keys()))
```

Вызовем функцию `train_test_split` для наших данных. 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)
```

По умолчанию 75\% идет в обучающий набор, 25\% - в тестовый. Перед разбиением функция `train_test_split` перемешивает набор данных с помощью генератора псевдослучайных чисел. Если мы просто возьмем последние 25\% наблюдений в качестве тестового набора, все
точки данных будет иметь метку 2, поскольку данные отсортированы по меткам (см. `iris['target']`). 

Чтобы сделать результат воспроизводимым, мы используем генератором псевдослучайных чисел с фиксированной базой генерации, которая задается с помощью параметра `random_state`. 

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
```

```python
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

Для визуализации данных используем диаграммы рассеяния для признаков обучающего набора. Точки данных окрашены в соответствии с сортами ириса, к которым они относятся. Чтобы построить диаграммы, мы сначала преобразуем массив `NumPy` в `DataFrame`.

```python
# create dataframe from data in X_train
# label the columns using the strings in iris_dataset.feature_names
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
# create a scatter matrix from the dataframe, color by y_train
pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15),
                           marker='o', hist_kwds={'bins': 20}, s=60,
                           alpha=.8);
```

Графики показывают, что признаки позволяют хорошо разделять метки. Это означает, что данная задача классификации достаточно проста.

Применим метод $k$ ближайших соседей ($k$-NN: $k$ Nearest Neighbors). Для предсказания метки, соответствующей вектору признаков $k$, данный метод берет $k$ ближайших векторов признаков из обучающего набора и проводит голосовании с помощью их меток. 

В `scikit_learn` данный алгоритм реализован в классификаторе `KNeighborsClassifier` модуля `neighbors`. Прежде чем использовать эту модель, нам нужно создать объект-экземпляр класса. Это произойдет, когда мы зададим параметры модели. Гиперпараметром `KNeighborsClassifier` является количество соседей, которые мы положим равным 1:

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

Объект `knn` включает в себя алгоритм, который будет использоваться для построения модели на обучающих данных, а также алгоритм, который сгенерирует прогнозы для новых данных. Для построения модели на обучающем наборе вызовем метод `fit` объекта `knn`, который принимает в качестве аргументов массивы `X_train, y_train`,  содержащие обучающие данные и их метки и возвращает измененный объект `knn`.

```python
knn.fit(X_train, y_train)
```

Предскажем метку нового экземпляра, которого не было в обучающем наборе.

```python
X_new = np.array([[5, 2.9, 1, 0.2]])
print("X_new.shape:", X_new.shape)
```

```python
prediction = knn.predict(X_new)
print("Prediction:", prediction)
print("Predicted target name:",
       iris_dataset['target_names'][prediction])
```

Модель предсказывает, что этот новый цветок ириса принадлежит к классу `0`, что означает сорт `setosa`. Для оценки качества модели предскажем метки тестового набора:

```python
y_pred = knn.predict(X_test)
print("Test set predictions:\n", y_pred)
```

и вычислим правильность (accuracy) - долю правильных предсказаний с помощью метода `score` объекта `knn`:

```python
print("Test set score: {:.2f}".format(knn.score(X_test, y_test)))
```

Это равносильно

```python
print("Test set score: {:.2f}".format(np.mean(y_pred == y_test)))
```
