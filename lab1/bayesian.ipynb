# **Импорты**

```python
import numpy as np
import matplotlib.pyplot as plt
import pylab

from numpy import linalg as la
```

# **Подготовка тренировочных и тестовых данных**

```python
train_size = 200
test_size = 80

x1_train = np.random.normal((3, 5), 0.8, size=(train_size, 2))
y1_train = np.full((train_size, 1), 1)

x2_train = np.random.normal((1, -1), 0.7, size=(train_size, 2))
y2_train = np.full((train_size, 1), 2)

x1_test = np.random.normal((3, 5), 0.8, size=(test_size, 2))
y1_test = np.full((test_size, 1), 1)

x2_test = np.random.normal((1, -1), 0.7, size=(test_size, 2))
y2_test = np.full((test_size, 1), 2)
```

```python
c = {1:'lime', 2:'magenta', 3:'green', 4:'m'}

plt.scatter(x1_train[:, 0], x1_train[:, 1], c=[c[i] for i in y1_train[:, 0]])
plt.scatter(x2_train[:, 0], x2_train[:, 1], c=[c[i] for i in y2_train[:, 0]])
plt.scatter(x1_test[:, 0], x1_test[:, 1], c=[c[i+2] for i in y1_test[:, 0]])
plt.scatter(x2_test[:, 0], x2_test[:, 1], c=[c[i+2] for i in y2_test[:, 0]])

pylab.ylim (-4, 8)
pylab.xlim (-4, 8)
```

# **Поиск оптимальных значений для каждого класса**

## **Выборочное среднее**

```python
m1 = (1 / x1_train.shape[0]) * np.sum(x1_train, axis=0)
m1
```

```python
m2 = (1 / x2_train.shape[0]) * np.sum(x2_train, axis=0)
m2
```

## **Выборочная ковариационная матрица**

```python
x1_m1_matrix = x1_train - m1
C1 = (1 / x1_train.shape[0]) * np.dot(x1_m1_matrix.T, x1_m1_matrix)
C1
```

```python
C1_inv = la.pinv(C1)
C1_det = C1[0][0] * C1[1][1] - C1[0][1] * C1[1][0]
C1_det
```

```python
x2_m2_matrix = x2_train - m2
C2 = (1 / x2_train.shape[0]) * np.dot(x2_m2_matrix.T, x2_m2_matrix)
C2 == C2.T
```

```python
C2_inv = la.pinv(C2)
C2_det = C2[0][0] * C2[1][1] - C2[0][1] * C2[1][0]
C2_det
```

# **Реализация решающего правила**

```python
def P1(x):
    p = (1 / np.sqrt(4 * np.pi * np.pi * C1_det)) * np.exp((-1/2) * np.dot(C1_inv, ((x - m1) * (x - m1)).T))
    return p
```

```python
def P2(x):
    p = (1 / np.sqrt(4 * np.pi * np.pi * C2_det)) * np.exp((-1/2) * np.dot(C2_inv, ((x - m2) * (x - m2)).T))
    return p
```

```python
def L(x):
    return np.mean(P1(x), axis=0) / np.mean(P2(x), axis=0)
```

# **Проверка на тестовой выборке**

```python
y1_pred = L(x1_test)
y2_pred = L(x2_test)

y1_pred[y1_pred < 1] = -100
y1_pred[y1_pred > 1] = 1

y2_pred[y2_pred > 1] = 100
y2_pred[y2_pred < 1] = 2

c = {1:'green', 2:'m', -100: 'black', 100:'black'}

plt.scatter(x1_test[:,0], x1_test[:,1], c=[c[y] for y in y1_pred])
plt.scatter(x2_test[:,0], x2_test[:,1], c=[c[y] for y in y2_pred])

pylab.ylim (-4, 8)
pylab.xlim (-4, 8)

plt.show()
```
